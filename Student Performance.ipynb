# DATA VISUALIZATION


# 1. We will see visualization based on gender with the help of bar graphs.

```python
sns.barplot(x=df['gender'].value_counts().index,y=df['gender'].value_counts().values)
plt.title('Genders other rate')
plt.ylabel('Rates')
#plt.legend(loc=0)
plt.show()
```

# 2. Count of the column race/ethnicity is shown with the below graph¶

```python
plt.figure(figsize=(7,7))
sns.barplot(x=df['race/ethnicity'].value_counts().index,
              y=df['race/ethnicity'].value_counts().values)
plt.xlabel('Race/Ethnicity')
plt.ylabel('Frequency')
plt.title('Show of Race/Ethnicity Bar Plot')
plt.show()
```

# 3. Plotting graphs showing comparison for ‘writing score’, ‘reading score’ and ‘math score’ for both the ‘genders’ based on ‘parental level of education’ respectively.

```python
plt.figure(figsize=(10,7))
sns.barplot(x = "parental level of education", y = "writing score", hue = "gender", data = df)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x = "parental level of education", y = "reading score", hue = "gender", data = df)
plt.xticks(rotation=45)
plt.show()
```

# 4. Plotting a graph for both the genders comparing math score for different levels of education.

```python
plt.figure(figsize=(12,6))
sns.catplot(y="gender", x="math score",
                 hue="parental level of education",
                 data=df,Kfind="bar")
plt.title('for Parental Level Of Education Gender & Math_Score')
plt.show()
```

# 5. Visualizing different groups based on percentage with the help of a piechart.

```python
labels=df['race/ethnicity'].value_counts().index
colors=['cyan','pink','orange','lightgreen','yellow']
explode=[0,0,0.1,0,0]
values=df['race/ethnicity'].value_counts().values

#visualization
plt.figure(figsize=(7,7))
plt.pie(values,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%')
plt.title('Race/Ethnicity According Analysis',color='black',fontsize=10)
plt.show()
```

# 6. Plotting a graph for math score vs writing score for both the genders using scatter plot.

```python
sns.lmplot(x='math score',y='writing score',hue='gender',data=df)
plt.xlabel('Math Score')
plt.ylabel('Writing Score')
plt.title('Math Score vs Writing Score')
plt.show()
```

# 7. Visualizing frequency of math score vs writing score vs reading score using kde plot.

```python
sns.kdeplot(df['math score'], shade=True,color='c')
sns.kdeplot(df['reading score'],shade=True,color='r')
sns.kdeplot(df['writing score'],shade=True,color='b')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Math Score vs Reading Score vs Writing Score Kde Plot System Analysis')
plt.show()
```

# 8. Visualization for math score for both the genders using Box Plot¶

```python
sns.boxplot(x=df['gender'],y=df['math score'])
plt.show()
```

# 9. Data visualization using Pairplot.

```python
sns.pairplot(df)
plt.show()
```
